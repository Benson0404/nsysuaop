```python
import pandas as pd

data = {
    'ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Helen', 'Ivy', 'Jack'],
    'Age': [25, 30, 35, 40, 22, 29, 38, 27, 33, 31],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose'],
    'Salary': [50000, 60000, 75000, 80000, 45000, 62000, 72000, 58000, 69000, 71000],
    'Department': ['IT', 'HR', 'Finance', 'Marketing', 'IT', 'HR', 'Finance', 'Marketing', 'IT', 'Finance']
}
df = pd.DataFrame(data)
df
```

## 1.將薪水高於第3四分位數的人員標示為高薪

```python
high_salary = df['Salary'].quantile(0.75)
high_salary
```

```python
df['High Salary'] = df['Salary'].apply(lambda x: 'Yes' if x > high_salary else 'No')
df
```

## 2.刪除高薪欄位

```python
df.drop(columns=['High Salary'], inplace=True)
df
```

## 3.使用 pd.concat() 新增資料（資料集）

```python
new_data = [
    {'ID': 111, 'Name': 'Kevin', 'Age': 28, 'City': 'Austin', 'Salary': None, 'Department': 'IT'},
    {'ID': 112, 'Name': 'Lily', 'Age': 26, 'City': 'Seattle', 'Salary': 55000, 'Department': 'Marketing'},
    {'ID': 112, 'Name': 'Lily', 'Age': 26, 'City': 'Seattle', 'Salary': 55000, 'Department': 'Marketing'}
]
new_df = pd.DataFrame(new_data)
new_df
```

```python
df = pd.concat([df, new_df], ignore_index=True)
df
```

## 4.找出空值的資料列，並填充空值（薪水依該部門平均值）

```python
df.isnull().sum()
```

```python
avg_dp = df.groupby('Department')['Salary'].mean()
avg_dp
```

```python
for i in range(len(df)):
  if pd.isnull(df.loc[i, 'Salary']):
    df.loc[i, 'Salary'] = avg_dp[ df.loc[i, 'Department'] ]

df
```

## 5.刪除重複資料

```python
df.drop_duplicates(inplace=True)
df
```

## 6.新增欄位 'NewSalary'，如果薪水小於平均值則薪水加 500，否則維持原薪水；新增欄位 'RankSalary' 填入薪水排名

```python
salary_avg = df['Salary'].mean()
salary_avg
```

```python
df['NewSalary'] = df['Salary'].apply(lambda x: x+500 if x<salary_avg else x)
df
```

```python
df['RankSalary'] = df['Salary'].rank(ascending=False)

# 數值轉換
df[['Salary','NewSalary','RankSalary']] = df[['Salary','NewSalary','RankSalary']].astype(int)
df
```

## 引言

### 背景
肥胖（Obesity）是當代全球公共衛生領域中最嚴峻的健康挑戰之一。根據世界衛生組織（WHO）的統計資料，全球肥胖人口數量在過去數十年間迅速上升，並成為多種慢性疾病（如第二型糖尿病、心血管疾病、某些癌症等）的主要危險因子。肥胖的成因十分複雜，除遺傳因素外，個人的生活習慣亦扮演著關鍵角色，包括飲食行為、身體活動量、作息模式以及其他日常行為。

隨著現代社會的都市化、科技化，人們的生活型態出現顯著變化：高熱量食物的攝取普及化、日常活動量減少、久坐時間延長等因素，皆可能促進肥胖的發生。同時，家族疾病史（如家族性肥胖、糖尿病史）也被認為與個人肥胖風險密切相關。因此，探討不同生活習慣因素與肥胖之間的關係，有助於深入了解肥胖形成的機制，並提供更具針對性的預防與干預建議。

### 目的
本研究旨在透過系統性分析多項生活習慣變數（如高熱量攝取頻率、每日蔬菜攝取頻率、運動頻率、飲水量、餐間進食習慣、主要交通方式等），以及個人背景（如家族肥胖史、糖尿病家族史）與肥胖分類（NObeyesdad）之間的關聯性，期望能找出影響肥胖的重要因素，並提供未來健康促進與肥胖防治計畫之參考依據。

## 結果與分析

1. 基礎描述性分析

```python
# 抽取需要分析的欄位
df_selected = df[['Age', 'Gender', 'Weight', 'NObeyesdad']]

# 計算分布
gender_counts = df_selected['Gender'].value_counts()
nobeyesdad_counts = df_selected['NObeyesdad'].value_counts()


# 1. 性別分布長條圖
plt.figure(figsize=(6, 4))
gender_counts.plot(kind='bar')
plt.title('性別分布 (Gender)')
plt.xlabel('性別')
plt.ylabel('人數')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 2. 肥胖分類分布長條圖
plt.figure(figsize=(8, 5))
nobeyesdad_counts.plot(kind='bar')
plt.title('肥胖分類分布 (NObeyesdad)')
plt.xlabel('肥胖分類')
plt.ylabel('人數')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 3. 年齡分布直方圖
plt.figure(figsize=(6, 4))
plt.hist(df_selected['Age'], bins=20, edgecolor='black')
plt.title('年齡分布 (Age)')
plt.xlabel('年齡')
plt.ylabel('人數')
plt.tight_layout()
plt.show()

# 4. 體重分布直方圖
plt.figure(figsize=(6, 4))
plt.hist(df_selected['Weight'], bins=20, edgecolor='black')
plt.title('體重分布 (Weight)')
plt.xlabel('體重 (kg)')
plt.ylabel('人數')
plt.tight_layout()
plt.show()
```

2. 生活習慣與肥胖的關係
  *   家族中有肥胖歷史的人 (family_history_with_overweight) 是否更容易肥胖？
  *   經常攝取高熱量食物 (FAVC) 與肥胖程度之間有顯著關係嗎？
  *   每日蔬菜攝取頻率 (FCVC) 較高的人，肥胖情況如何？
  *   進食次數較多 (NCP) 是否肥胖比例較高
  *   運動時間較多 (FAF) 是否肥胖比例較低
  *   進食次數 (NCP) 與運動時間 (FAF) 組合，對肥胖程度有何影響？
  *   在正餐之間是否有吃其他東西 (CAEC) 與肥胖程度之間有顯著關係嗎？
  *   是否有抽菸 (SMOKE) 與肥胖程度之間有顯著關係嗎？
  *   常喝水 (CH2O高的人) 是否肥胖比例較低？
  *   是否會監控自身攝取的卡路里 (SCC) 與肥胖程度之間有顯著關係嗎
  *   使用電子產品時間較多 (TUE) 是否肥胖比例較高
  *   喝酒頻率較高 (CALC) 是否肥胖比例較高
  *   搭乘交通工具方式 (MTRANS) 是否會影響肥胖率？（例如，步行 vs 公共交通 vs 私人汽車）






```python
import pandas as pd
import matplotlib.pyplot as plt

# 生活習慣變數清單
lifestyle_columns = [
    'family_history_with_overweight', 'FAVC', 'FCVC', 'FAF', 'NCP',
    'CAEC', 'SMOKE', 'CH2O', 'SCC', 'TUE', 'CALC', 'MTRANS'
]

# 肥胖程度欄位
target = 'NObeyesdad'

# 指定NObeyesdad順序
obesity_order = [
    'Insufficient_Weight', 'Normal_Weight',
    'Overweight_Level_I', 'Overweight_Level_II',
    'Obesity_Type_I', 'Obesity_Type_II', 'Obesity_Type_III'
]

# 把NObeyesdad轉成有順序的類別型資料
df[target] = pd.Categorical(df[target], categories=obesity_order, ordered=True)

# 繪製每個生活變數與肥胖分類的關係
for col in lifestyle_columns:
    plt.figure(figsize=(10, 5))

    if df[col].dtype == 'object':  # 類別型資料
        cross_tab = pd.crosstab(df[col], df[target], normalize='index')
        cross_tab = cross_tab[obesity_order]  # 手動調整欄位順序
        cross_tab.plot(kind='bar', stacked=True)
        plt.title(f'{col} 與 {target} 的關係')
        plt.xlabel(col)
        plt.ylabel('比例')
        plt.legend(title='NObeyesdad', bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.tight_layout()
        plt.show()

    else:  # 數值型資料（連續變數）
        # 只加 patch_artist=True 讓盒狀圖有顏色
        df.boxplot(column=col, by=target, figsize=(8,5), patch_artist=True)
        plt.title(f'{col} 與 {target} 的關係')
        plt.suptitle('')
        plt.xlabel('NObeyesdad')
        plt.ylabel(col)
        plt.xticks(ticks=range(1, len(obesity_order)+1), labels=[
            'Insufficient\nWeight', 'Normal\nWeight',
            'Overweight\nLevel I', 'Overweight\nLevel II',
            'Obesity\nType I', 'Obesity\nType II', 'Obesity\nType III'
        ])
        plt.tight_layout()
        plt.show()
```

🔹 家族肥胖史（family_history_with_overweight）

有家族肥胖史者（"yes"）肥胖風險大幅增加。

其中 Obesity Type II 和 Obesity Type III 的比例明顯高於無家族史者。

說明：基因遺傳及家庭環境（如飲食習慣）共同增加肥胖發生率。


---



🔹 高熱量飲食習慣（FAVC）

經常食用高熱量食物（"yes"）者，肥胖程度明顯提升。

特別是 Obesity Type II 和 III 的人數比例大幅增加。

說明：持續高熱量攝取會促使體脂肪累積，增加肥胖風險。


---



🔹 每日蔬菜攝取頻率（FCVC）

蔬菜攝取頻率（FCVC值）越高的人，體重控制越好。

Insufficient Weight 和 Normal Weight 的人，FCVC平均值明顯高於肥胖群體。

說明：高纖維食物能增加飽足感、降低總熱量攝取，有助於體重控制。

---

🔹 每週運動頻率（FAF）

每週運動頻率（FAF值）較高者（如每週運動2~3次以上），肥胖比例顯著降低。

Obesity Type III 群體運動頻率普遍偏低。

說明：規律運動能增加能量消耗，幫助維持健康體重。

---

🔹 每日進食次數（NCP）

進食次數適中（每天2~3次）的人，肥胖風險較低。

過多或過少進食次數（如每天>4次或僅1次）的人，肥胖風險反而增加。

說明：進食頻率失衡可能導致總熱量過量或代謝異常，影響體重。

---

🔹 餐間吃其他食物（CAEC）

餐間「Always」或「Sometimes」進食者，肥胖程度偏高。

"no" 的族群中 Normal Weight 比例較高。

說明：頻繁的餐間加餐可能增加額外熱量攝取，導致體脂肪累積。

---

🔹 抽菸習慣（SMOKE）

吸菸者肥胖比例稍低，但 Obesity Type II 的比例仍偏高。

說明：吸菸與體重間有複雜交互作用，部分吸菸者因代謝率提高而體重偏低，但同時也可能伴隨不健康生活模式，增加其他慢性病風險。

---

🔹 每日飲水量（CH2O）

飲水量高（每天>2公升）者，Normal Weight 比例較高。

肥胖族群飲水量普遍較低。

說明：充足飲水有助於新陳代謝及食慾控制，對維持正常體重有正面影響。

---

🔹 糖尿病家族史（SCC）
有糖尿病家族史（"yes"）者肥胖風險更高。

特別是 Obesity Type II、III 的比例上升。

說明：糖尿病和肥胖同屬代謝疾病，有共同的遺傳與生活型態因素。

---

🔹 使用電子產品時間（TUE）

長時間使用電子產品者，肥胖程度略有增加。

TUE時間分布範圍廣，推測靜態生活型態為肥胖危險因子之一。

說明：長時間坐著使用電子產品，活動量下降，熱量消耗減少，容易導致能量失衡。

---

🔹 飲酒頻率（CALC）

Always 飲酒者以 Normal Weight 為主。

反而「有時候喝」或「不喝」的人，肥胖比例增加（可能因生活型態差異造成）。

說明：單純飲酒頻率與肥胖間的關係複雜，需同時考量總飲食型態與活動量。

---

🔹 主要交通方式（MTRANS）

使用私家車（Automobile）通勤者，肥胖比例明顯高於步行（Walking）或騎自行車（Bike）的人。

步行族群以 Normal Weight 為主，肥胖比例最低。

說明：交通方式影響每日身體活動量，間接影響體重控制。





### 結論
1. 高熱量攝取、低運動頻率、低蔬菜攝取量、少喝水、偏好汽車通勤等因素均與肥胖程度呈正向關聯。
2. 家族遺傳背景（肥胖史、糖尿病史）亦是重要的肥胖危險因子。
3. 部分因素（如飲酒、吸菸）與肥胖關係較為複雜，需結合生活型態整體分析。

### 建議
1. 衛生機構應針對高風險族群（有家族病史、生活型態不良者）設計針對性健康促進方案。
2. 透過社區介入，強化民眾對**飲食管理、運動推廣、健康交通選擇（如步行、自行車）**的意識。
3. 提倡適量飲水、規律作息、減少靜態行為（如長時間用手機或電腦），降低肥胖與慢性病風險。

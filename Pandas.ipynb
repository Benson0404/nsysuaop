## Series

```python
# 載入Pandas模組
import pandas as pd

# 宣告建立一維資料 Series
age = [25, 30, 35, 40, 22, 29, 38, 27, 33, 31]
se = pd.Series(age)
se
```

```python
se[3]

#用索引值名稱
se.loc[3]

#用索引值編號
se.iloc[3]
```

```python
# 查看資料值
se.values

# 查看索引
se.index

# 查看資料型態
se.dtype
```

```python
# 自訂索引
name = ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Helen', 'Ivy', 'Jack']
age = [25, 30, 35, 40, 22, 29, 38, 27, 33, 31]
seidx = pd.Series(age, index=name)
seidx
```

##DataFrame

```python
#建立DataFrame
import pandas as pd

data = {
    'ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Helen', 'Ivy', 'Jack'],
    'Age': [25, 30, 35, 40, 22, 29, 38, 27, 33, 31],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose'],
    'Salary': [50000, 60000, 75000, 80000, 45000, 62000, 72000, 58000, 69000, 71000],
    'Department': ['IT', 'HR', 'Finance', 'Marketing', 'IT', 'HR', 'Finance', 'Marketing', 'IT', 'Finance']
}
df = pd.DataFrame(data)
df
```

```python
# 新增一個「公司名稱」欄位，所有值都為「TechCorp」
df['Company'] = 'TechCorp'
df
```

```python
# 新增「年薪」欄位（假設每年有 12 個月薪資）
df['Annual Salary'] = df['Salary'] * 12
df
```

```python
# 新增「高薪」欄位（若薪資高於 65000 則標記為 Yes，否則為 No）
df['High Salary'] = df['Salary'].apply(lambda x: 'Yes' if x > 65000 else 'No')

def check_high_salary(x):
    if x > 65000:
        return 'Yes'
    else:
        return 'No'

df['High Salary 2'] = df['Salary'].apply(check_high_salary)

df
```

```python
# 刪除「City」與「Department」欄位
df.drop(columns=['High Salary 2'], inplace=True)
df
```

```python
df = df.drop(columns=['High Salary'])
df
```

```python
# 使用串列方式新增一筆資料（按照 DataFrame 欄位順序）
df.loc[len(df)] = [111, 'Kevin', 28, 'Austin', 67000, 'IT', 'TechCorp', '360005']
df
```

```python
new_data = [
    {'ID': 112, 'Name': 'Lily', 'Age': 26, 'City': 'Seattle', 'Salary': 55000, 'Department': 'Marketing'}
]
df = pd.concat([df, pd.DataFrame(new_data)], ignore_index=True)
df
```

```python
# 刪除索引為 2 的資料
df.drop(index=12, inplace=True) #inplace=True -> 會取代原本的dataframe
df
```

```python
# 刪除薪水高於 75000 的員工 -> 篩選
condition = df['Salary'] < 75000
df = df[condition]
df
```

```python
# 計算每一欄有多少個 NaN
df.isnull().sum()
```

```python
df.isnull()
```

```python
# 找出具有空值的列(回傳每一列是否「至少有一個欄位是 NaN」的布林值)
df.isnull().any(axis=1)

#只看 'ID' 和 'Name' 這兩欄
df[['ID','Name']].isnull().any(axis=1)

#直接把「至少有一個欄位是 NaN」的整列資料取出來
df[df.isnull().any(axis=1)]
```

```python
# 找出具有空值的欄
df.columns[df.isnull().any(axis=0)]
```

```python
df.loc[4,'Salary'] = pd.NA
df
```
